```python
import numpy as np
from nhlpy.api.teams import Teams
from yahoo_oauth import OAuth2
import yahoo_fantasy_api as yfa
from yahoo_fantasy_api import *
import pandas as pd
```

```python
# Authenticate using credentials from oauth2.json
sc = OAuth2(None, None, from_file='C:/Users/ryand/OneDrive/Documents/fantasy_lineup_optimizer/basic-app/oauth2.json')

sc.refresh_access_token()
```

```python
game = yfa.Game(sc, 'nhl')
league_ids = game.league_ids(year=2025)


# Load a league object and print its standings
league = game.to_league(league_ids[3])
standings = league.standings()

standings_df = pd.json_normalize(standings)

# print("Standings:", standings)
teams = league.teams()
teams
```

```python
import pandas as pd

# Flatten the team dictionary into a list of dicts for tabular conversion
team_list = []
for team in teams.values():
    manager = team["managers"][0]["manager"] if team["managers"] else {}
    team_list.append({
        "team_name": team["name"],
        "team_id": team["team_id"],
        "team_key": team["team_key"],
        "url": team["url"],
        "logo_url": team["team_logos"][0]["team_logo"]["url"] if team["team_logos"] else "",
        "waiver_priority": team["waiver_priority"],
        "number_of_moves": team["number_of_moves"],
        "number_of_trades": team["number_of_trades"],
        "roster_adds": team["roster_adds"]["value"] if "roster_adds" in team else "",
        "scoring_type": team["league_scoring_type"],
        "has_draft_grade": team["has_draft_grade"],
        "manager_nickname": manager.get("nickname", ""),
        "manager_email": manager.get("email", ""),
        "manager_felo_score": manager.get("felo_score", ""),
        "manager_felo_tier": manager.get("felo_tier", ""),
    })

# Create DataFrame and sort by team name
df = pd.DataFrame(team_list)
df_sorted = df.sort_values("team_key")

team_keys = df_sorted['team_key']
team_keys
```
